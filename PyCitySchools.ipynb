```python
# Dependencies and Setup
import pandas as pd
import os
# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school = pd.read_csv(school_data_to_load)
student = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
complete = pd.merge(student, school, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school=len(complete['school_name'].unique())
student=complete['student_name'].count()
df=pd.DataFrame(complete['budget'].unique())
budget=df[0].sum()
math=complete['math_score'].mean()
read=complete['reading_score'].mean()
passing=(math+read)/2
passing_math=(len(complete[complete['math_score']>=70]))/student*100
passing_read=(len(complete[complete['reading_score']>=70]))/student*100
district=pd.DataFrame({'Total Schools':school, 'Total Students':student,'Total Budget': budget,'Average Math Score':math,'Average Reading Score':read,'% Passing Math':passing_math,'% Passing Reading':passing_read,'% Overall Passing Rate':passing}, index=[0])
district
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
schools=complete.groupby(['school_name'])
sch_summary=pd.DataFrame(schools['type'].first())
sch_summary['Total Students']=schools['Student ID'].count()
sch_summary['Total School Budget']=schools['budget'].first()
sch_summary['Per Student Budget']=schools['budget'].first()/schools['Student ID'].count()
sch_summary['Average Math Score']=schools['math_score'].mean()
sch_summary['Average Reading Score']=schools['reading_score'].mean()
```

```python
math_70=complete[complete['math_score']>=70]
read_70=complete[complete['reading_score']>=70]
a=math_70.groupby(['school_name'])
b=read_70.groupby(['school_name'])
```

```python
sch_summary['% Passing Math']=(a['Student ID'].count())/(schools['Student ID'].count())
sch_summary['% Passing Reading']=(b['Student ID'].count())/(schools['Student ID'].count())
sch_summary['% Overall Passing Rate']=(sch_summary['% Passing Math']+sch_summary['% Passing Reading'])/2
sch_summary
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top=sch_summary.sort_values('% Overall Passing Rate', ascending=False)
top.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom=sch_summary.sort_values('% Overall Passing Rate', ascending=True)
bottom.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
nineth=complete[complete['grade']=='9th']
tenth=complete[complete['grade']=='10th']
eleventh=complete[complete['grade']=='11th']
twelveth=complete[complete['grade']=='12th']
nine=nineth.groupby('school_name')
nine_math=pd.Series(nine['math_score'].mean())
ten=tenth.groupby('school_name')
ten_math=pd.Series(ten['math_score'].mean())
eleven=eleventh.groupby('school_name')
eleven_math=pd.Series(eleven['math_score'].mean())
twelve=twelveth.groupby('school_name')
twelve_math=pd.Series(twelve['math_score'].mean())
mathsummary=pd.concat([nine_math,ten_math,eleven_math,twelve_math], axis=1)
# mathsummary=mathsummary.rename(columns={'math_score':'9th','math_score':'10th','math_score':'11th','math_score':'12th'})
mathsummary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
sch_summary['Spending Ranges']=pd.cut(sch_summary['Per Student Budget'], spending_bins, labels=group_names)
sch_summary.head()
```

```python
sch=sch_summary.groupby('Spending Ranges')
school_spend=pd.DataFrame(sch['Average Math Score'].mean())
school_spend['Average Reading Score']=sch['Average Reading Score'].mean()
school_spend['% Passing Math']=sch['% Passing Math'].mean()*100
school_spend['% Passing Reading']=sch['% Passing Reading'].mean()*100
school_spend['% Overall Passing Rate']=(school_spend['% Passing Math']+school_spend['% Passing Reading'])/2
school_spend
```
